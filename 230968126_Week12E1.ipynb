1.
How many transactions and items are there in the data set?

```python
with open('groceries.csv', 'r') as f:
    lines = f.readlines()
data = [line.strip().split(',') for line in lines]
max_items = max(len(row) for row in data)
column_names = ['Item' + str(i) for i in range(1, max_items + 1)]
df = pd.DataFrame(data, columns=column_names)
df = df.replace('', np.nan)
df
```

```python
num_transactions = len(df)
num_items = df.count().sum()
print(f"Number of transactions: {num_transactions}")
print(f"Number of items: {num_items}")
```

2.
Prepare the data for finding association rules. Each transaction will contain a list of item in the transaction. [['citrus fruit', 'semi-finished bread', 'margarine', 'ready soups'], ['tropical fruit', 'yogurt', 'coffee'],……. ['whole milk']]

```python
transactions = []
for _, row in df.iterrows():
  transaction = [item for item in row if pd.notna(item)]
  transactions.append(transaction)

transactions
```

3. Use Python library mlxtend and convert the transactions into a format that can be used in the Apriori method for finding frequent itemsets. pip install mlxtend from mlxtend.preprocessing import TransactionEncoder from mlxtend.frequent_patterns import apriori, association_rules

```python
#!pip install mlxtend

from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori, association_rules

te = TransactionEncoder()
te_ary = te.fit(transactions).transform(transactions)
df_trans = pd.DataFrame(te_ary, columns=te.columns_)
df_trans
```

4.
Find top selling items with minimum support of 2%.

```python
frequent_itemsets = apriori(df_trans, min_support=0.02, use_colnames=True)
frequent_itemsets
```

5.
Find all frequent itemsets with minimum support of 5%.

```python
frequent_itemsets_5 = apriori(df_trans, min_support=0.05, use_colnames=True)
frequent_itemsets_5
```

6.
Find all frequent itemsets of length 2 with minimum support of 2%.

```python
frequent_itemsets_length_2 = frequent_itemsets[frequent_itemsets['itemsets'].apply(lambda x: len(x)) == 2]
frequent_itemsets_length_2
```

7.
Find the top 10 association rules with minimum support of 2%, sorted by confidence in descending order.

```python
rules = association_rules(frequent_itemsets, metric="confidence", min_threshold=0.02)
top_10_rules = rules.sort_values('confidence', ascending=False).head(10)
top_10_rules
```

8.
Find association rules with minimum support of 2% and lift of more than 1.0.

```python
rules_lift = association_rules(frequent_itemsets, metric="lift", min_threshold=1.0)
rules_lift_filtered = rules_lift[(rules_lift['support'] >= 0.02) & (rules_lift['lift'] > 1.0)]
rules_lift_filtered
```
