1.
Read about the movielens dataset and write down a summary of metadata.

```python
links_df = pd.read_csv('links.csv')
movies_df = pd.read_csv('movies.csv')
ratings_df = pd.read_csv('ratings.csv')
tags_df = pd.read_csv('tags.csv')


movielens_df = pd.merge(movies_df, ratings_df, on='movieId', how='inner')
movielens_df = pd.merge(movielens_df, tags_df, on='movieId', how='inner')
movielens_df = pd.merge(movielens_df, links_df, on='movieId', how='inner')
movielens_df
```

```python
movielens_df = movielens_df.loc[:,~movielens_df.columns.duplicated()]
movielens_df.dropna(inplace=True)
metadata = {
    'Number of movies': movielens_df['movieId'].nunique(),
    'Number of unique genres': movielens_df['genres'].nunique(),
    'Number of ratings': movielens_df.shape[0],
    'Average rating': movielens_df['rating'].mean(),
    'Number of unique tags': movielens_df['tag'].nunique(),
}

metadata
```

2.
Read the “ratings.csv” file and create a pivot table with index=‘userId’, columns=‘movieId’, values = “rating.


```python
ratings_pivot = ratings_df.pivot_table(index='userId', columns='movieId', values='rating')
ratings_pivot
```

3.
sklearn.metrics.pairwise_distances can be used to compute distance between all pairs of users. pairwise_distances() takes a metric parameter for what distance measure to use. Use cosine similarity for finding similarity among users. Use the following packages.
from sklearn.metrics import pairwise_distances
from scipy.spatial.distance import cosine, correlation

```python
from sklearn.metrics import pairwise_distances
from scipy.spatial.distance import cosine, correlation

user_similarity = 1 - pairwise_distances(ratings_pivot.fillna(0), metric='cosine')
user_similarity_df = pd.DataFrame(user_similarity, index=ratings_pivot.index, columns=ratings_pivot.index)
user_similarity_df
```

4.
Find the 5 most similar user for user with user Id 10.

```python
user_id = 10
similar_users = user_similarity_df[user_id].sort_values(ascending=False)[1:6]

print("Top 5 most similar users for user", user_id, "are:")
similar_users
```

5.
Use the “movies” dataset to find out the names of movies, user 2 and user 338 have watched in common and how they have rated each one of them.

```python
user2_ratings = ratings_df[ratings_df['userId'] == 2]
user338_ratings = ratings_df[ratings_df['userId'] == 338]

common_movies = pd.merge(user2_ratings, user338_ratings, on='movieId')

common_movies_with_names = pd.merge(common_movies, movies_df, on='movieId')

print("Movies watched in common by user 2 and user 338:")
print(common_movies_with_names[['title', 'rating_x', 'rating_y']])
```

6.
Use the movies dataset to find out the common movie names between user 2 and user 338 with least rating of 4.0

```python
common_movies_high_rating = common_movies_with_names[
    (common_movies_with_names['rating_x'] >= 4.0) &
    (common_movies_with_names['rating_y'] >= 4.0)
]

print("Common movies with a least rating of 4.0 by user 2 and user 338:")
print(common_movies_high_rating[['title', 'rating_x', 'rating_y']])
```

7.
Create a pivot table for representing the similarity among movies using correlation

```python
movie_similarity = ratings_pivot.corr(method='pearson')
movie_similarity
```

8.
Find the top 5 movies which are similar to the movie “Godfather”.

```python
godfather_movie = movies_df[movies_df['title'].str.contains('Godfather')]
godfather_movie_id = godfather_movie['movieId'].values[0]
godfather_similarity_scores = movie_similarity[godfather_movie_id].dropna()
top_similar_movies = godfather_similarity_scores.sort_values(ascending=False)[1:6]
top_similar_movie_ids = top_similar_movies.index.tolist()
top_similar_movie_titles = movies_df[movies_df['movieId'].isin(top_similar_movie_ids)]['title'].tolist()
print("Top 5 movies similar to 'Godfather':")
for i, title in enumerate(top_similar_movie_titles):
  print(f"{i+1}. {title}")
```
