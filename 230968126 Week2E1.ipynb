2) For each of the numeric attributes (proteins up to vitamins), identify and replace all 
missing data (indicated with ?) with the arithmetic mean of the attribute.

```python
df2 = df1.replace("?",-1)
df2 = df2.mask(df2 == -1)
for i in df2.columns:
    if df2[i].isnull().any():
        df2[i] = df2[i].fillna(df2[i].astype(float).mean())
df2
```

4) For each of the numeric attributes (proteins up to vitamins), identify and replace all 
noisy data with the median of attribute.

```python
for column in df2.columns:

    if pd.api.types.is_numeric_dtype(df2[column]):

        median = df2[column].median()
        Q1 = df2[column].quantile(0.25)
        Q3 = df2[column].quantile(0.75)
        IQR = Q3 - Q1
        

        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR

        df2[column] = np.where(
            (df2[column] < lower_bound) | (df2[column] > upper_bound),
            median,
            df2[column]
        )


(df2)
```
