```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
houses = pd.read_csv("D:\\230968126\\Week2\\Bengaluru_House_Prices.csv")
df = pd.DataFrame(houses)
df
```

```python
missing_values = df.isnull().sum()
print("Missing values in each column:")
print(missing_values)
total_missing_values = df.isnull().sum().sum()
print("Total number of missing values is: ",total_missing_values)
print("Percentage of missing values in dataframe is: ",(total_missing_values/13320)*100)
```

```python
threshold = 0.15 * len(df)
columns_to_drop = [col for col in df.columns if df[col].isnull().sum() > threshold and col not in ['location', 'size', 'total_sqft']]
df = df.drop(columns=columns_to_drop)
print("Column with most missing values is society")
df
```

Here we removed society column as more than 15% of its values are missing

```python
df_cleaned = df.loc[:,["location","size","total_sqft"]].dropna(inplace=True)
df
```

```python
df = df[df['bath'] <= 5]
df
```

```python
house_counts = df['location'].value_counts()


house_counts_df = house_counts.reset_index()
house_counts_df.columns = ['Location', 'Number of Houses']


print("Number of houses available in each area:")
print(house_counts_df)
```

```python
print("Top 5 areas with most houses: ",house_counts_df.head())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

top_5_locations = df['location'].value_counts().nlargest(5).index

df_top_5 = df[df['location'].isin(top_5_locations)]


plt.figure(figsize=(15, 10))

for location in top_5_locations:
    plt.subplot(2, 3, top_5_locations.tolist().index(location) + 1)
    subset = df_top_5[df_top_5['location'] == location]
    sns.scatterplot(data=subset, x='total_sqft', y='price')
    plt.title(f'Price vs. Build Up Area in {location}')
    plt.xlabel('Build Up Area (sqft)')
    plt.ylabel('Price')
    plt.grid(True)

plt.tight_layout()
plt.show()
```
